```python
def count_total_of_even(numbers):
    total = 0
    for number in numbers:
        if number % 2 == 0:
            total += 1
    return total  

count_total_of_even(numbers)
```

```python
Nums = [12, 4, 27, 3, 188, 50, 7]

def find_largest(nums):
    if not nums:
        return "no values to compare"
    
    largest = nums[0]

    for n in nums[1:]:
        if n > largest:
            largest = n

    return largest

print(find_largest(Nums))
```
