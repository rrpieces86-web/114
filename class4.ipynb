```python
# Challenge 1:
# node
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

# linked list

class LinkedList:
    def __init__(self):
        self.head = None

    # append method
    def append(self, value):
        # add a new node at the end of the list
        new_node = Node(value)

        # if the list is empty new node becomes the head

        if self.head is None:
            self.head = new_node
            return
        # otherwise, walk to the end and attach it
        current = self.head
        while current.next is not None:
            current = current.next

        current.next = new_node
    

    # print_list method
    def print_list(self):
        current = self.head

        if current is None:
            print("list is empty")

        while current is not None:
            print(current.value)
            current = current.next

    def count_nodes(self):
        count = 0
        current = self.head
        while current is not None:
            count += 1
            current = current.next
        return count
    
    # insert at the beginning 
    def prepend(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node

    # remove method
    def remove(self, value):
        # remove the first node that has the given value.
        #if the list is empty, nothing to do
        if self.head is None:
            return
        
        # if the node to remove is the head
        if self.head.value == value:
            self.head= self.head.next
            return
        
        # otherwise search for the node before the one to remove
        current = self.head
        while current.next is not None:
            if current.next.value == value:
                # skip the node with the matching value
                current.next = current.next.next
                return
            current = current.next

# create a linked list and append values 10 15 20 15 30

sll = LinkedList()

sll.append(10)
sll.append(15)
sll.append(20)
sll.append(25)
sll.append(30)

# print list
print("Initial singly list (after append): ")
sll.print_list()
sll.count_nodes()
sll.prepend(50)
sll.print_list()
sll.remove(25)
sll.print_list()

print("after remove: ")
sll.print_list()
```

```python
# node class
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

# doubly linked lis
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # append method
    def append(self, value):
        # create a new node with the given value
        new_node = Node(value)
        # check if the list is empty
        if self.head is None:
            # if it is empty make the new node the head
            self.head = new_node
            #stop
            return
        # start traversing the list from the head
        current = self.head
        # move forward until we reach the last node
        while current.next is not None:
            current = current.next

        # link the last node to the new node
        current.next = new_node

        # link the new node back to the previous last node
        new_node.prev = current

    # Print forward method
    def print_forward(self):
        current = self.head

        if current is None:
            print("list is empty")
            return
        
        while current is not None:
            print(current.value)
            current = current.next

    # print  backwards method
    def print_backward(self):
        current = self.head

        if current is None:
            print("list is empty")
            return
        
        # move to tail
        while current.next is not None:
            current = current.next

        while current is not None:
            print(current.value)
            current = current.prev



# testing classes and methods
dll = DoublyLinkedList()

# add values
dll.append("A")
dll.append("B")
dll.append("C")
dll.append("D")

print("print forward: ")
dll.print_forward()
print("print backwards: ")
dll.print_backward()

    
```
