```python
# challenge 1: push and pop

# create empty stack
stack = []

# push values a, b, and c into the stack in order
stack.append("A")
stack.append("B")
stack.append("C")

# pop the top value and store it in a variable called top
top = stack.pop()

# print stack and value that was removed
print("stack after first pop:", stack)
print("removed value:", top)

# pop the next value and print it
removed = stack.pop()
print("next value:", removed)

# push value d into the stack
stack.append("D")

# pop again and print removed value
removed = stack.pop()
print("removed value:", removed)

#print stack
print("final stack:", stack)



```

```python
word = "hello"
stack = []
reverse_word = ""

for char in word:
    stack.append(char)

while len(stack) > 0:
    reverse_word += stack.pop()

print(reverse_word)
```

class 2 - stacks and queues

1. stacks - a stack is an ordered collection of elements where items are added and removed from one end, commonly referred to as the "top", the ordering principle is referred to is lifo (last in, first out)

2. queues - a queue is an ordered collection of elements where items and are added and removed from *opposite* ends, commonly referred to as top and bottom or rear of the queue, the ordering principle is referred to as fifo (first in, first out)
